```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
X=data.drop(['median_house_value'],axis=1)
y=data['median_house_value']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
td=X_train.join(y_train)
```

```python
td
```

```python
td.hist(figsize=(15,8))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(td.corr(),annot=True,cmap="YlGnBu")
```

```python
td['total_rooms']=np.log(td['total_rooms']+1)
td['total_bedrooms']=np.log(td['total_bedrooms']+1)
td['population']=np.log(td['population']+1)
td['households']=np.log(td['households']+1)
```

```python
td.hist(figsize=(15,8))
```

```python
td=td.join(pd.get_dummies(td.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

```python
td
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(td.corr(),annot=True,cmap="YlGnBu")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude",y="longitude",data=td,hue="median_house_value",palette="coolwarm")
```

```python
td['bedroom_ratio']=td['total_bedrooms']/td['total_rooms']
td['households_rooms']=td['total_rooms']/td['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(td.corr(),annot=True,cmap="YlGnBu")
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
X_train,y_train=td.drop(['median_house_value'],axis=1),td['median_house_value']
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
testd=X_test.join(y_test)
testd['total_rooms']=np.log(testd['total_rooms']+1)
testd['total_bedrooms']=np.log(testd['total_bedrooms']+1)
testd['population']=np.log(testd['population']+1)
testd['households']=np.log(testd['households']+1)
testd=testd.join(pd.get_dummies(testd.ocean_proximity)).drop(['ocean_proximity'],axis=1)
testd['bedroom_ratio']=testd['total_bedrooms']/testd['total_rooms']
testd['households_rooms']=testd['total_rooms']/testd['households']
X_test,y_test=testd.drop(['median_house_value'],axis=1),testd['median_house_value']
```

```python
testd
```

```python
reg.score(X_test,y_test)
```
